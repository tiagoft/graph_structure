# The Hypergeometric Lower Bound

Suppose we have a bag containing $n$ numbered balls, which are numbered from $1$ to $n$. We build set $A$ by randomly drawing $k$ balls from the bag without replacement. Then, we put all balls back in the ball and build set $B$ by drawing the same number $k$ of balls without replacement. The question is: what is the cardinality of the intersection between $A$ and $B$?

## Solution: hypergeometric distribution

After building set $A$, we can mark all balls with a cross before replacing them. Then, when building set $B$, we have the following situation:

* We have $n$ balls,
* There are $k$ balls of interest,
* There are $k$ draws without replacement
* $x = |A \cap B|$ corresponds to the number of successes within these draws.

This directly leads to a hypergeometric distribution:

$$
p(x; n, k) = \text{Hyp}(x,n,k,k) = \frac{{\binom{k}{x} \binom{n-k}{k-x}}}{{\binom{n}{k}}}
$$

And, following the properties of a [hypergeometric distribution](https://openstax.org/books/introductory-statistics/pages/4-5-hypergeometric-distribution):

$$
\mathbb{E}[x] = \frac{\text{\# objects of interest} \times \text{\# draws}}{\text{\# total objects}} = \frac{k^2}{n}
$$

## The case of Jaccard Similariy

The Jaccard Similarity between sets $A$ and $B$, with $k$ draws, is:

$$
J_s(A,B,k) = \frac{|A \cap B|}{|A \cup B|}.
$$

However, $|A \cup B| = |A| + |B| - |A \cap B| = 2k - |A \cap B|$.

Also, $|A \cap B|$ is a random variable $x$ and $\mathbb{E}[x] = \frac{k^2}{n}$.

Hence:
$$
\mathbb{E}[J_s(A,B,k)] = \frac{\mathbb{E}[x]}{2k-\mathbb{E}[x]} = \frac{k^2/n}{2k-(k^2/n)}.
$$

We can simplify:

$$
\frac{k^2/n}{2k-(k^2/n)} = \frac{k^2}{n(2k-(k^2/n))} = \frac{k^2}{2nk-k^2}.
$$

Since we know that $k$ is strictly positive ($k>0$), then:

$$
\mathbb{E}[J_s(A,B,k)] = \frac{k}{2n-k}
$$

## Choosing $k$ as a function of $M$

If $k=c n, c \in [0,1]$, then we have:

$$
\mathbb{E}[J_s(A,B,k)] = \frac{cn}{2n-cn} = \frac{cn}{(2-c)n} = \frac{c}{2-c}
$$

This leads to a curve like this:

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
import numpy as np

c = np.linspace(0,1,100)
ejs = c/(2-c)

plt.figure(figsize=(7,2))
plt.plot(c, ejs)
plt.xlabel('c = k/n')
plt.ylabel('$\mathbb{E}[J_s(A,B,k)]$')
plt.show()
```
